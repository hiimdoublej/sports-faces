```python
import requests

response = requests.get("https://api.motogp.com/riders-api/season/2023/riders")
riders = response.json()
```

```python
avatars = {}
for rider in riders:
    name = f'{rider["name"]} {rider["surname"]}'.strip().replace(' ', '-')
    link = rider['current_career_step']['pictures']['portrait']
    avatars[name] = link
```

```python
import os
output_dir = '../outputs/motogp'
os.makedirs(output_dir, exist_ok=True)

for name, link in avatars.items():
    response = requests.get(link)
    with open(os.path.join(output_dir, f'{name}.jpg'), 'wb') as fp:
        fp.write(response.content)
```
