```python
# curl  -H 'user-agent: Mozilla/5.0 (X11; Linux x86_64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/114.0.0.0 Safari/537.36'  
# -H 'x-api-key: 0TvQnueqKa5mxJntVWt0w4LpLfEkrV1Ta8rQBb9Z'  "https://esports-api.lolesports.com/persisted/gw/getLeagues?hl=zh-TW"

import requests
host = 'https://esports-api.lolesports.com'
headers = {
    'x-api-key': '0TvQnueqKa5mxJntVWt0w4LpLfEkrV1Ta8rQBb9Z'
}
params = {
    'hl': 'en-US'
}

leagues = requests.get(f'{host}/persisted/gw/getLeagues', params=params, headers=headers).json()['data']['leagues']
```

```python
# https://esports-api.lolesports.com/persisted/gw/getTournamentsForLeague?hl=zh-TW&leagueId=104366947889790212

latest_tournaments = []
for league in leagues:
    response = requests.get(f'{host}/persisted/gw/getTournamentsForLeague', params=params | {'leagueId': league['id']}, headers=headers)
    # print(response.json())
    latest_tournaments.append(response.json()['data']['leagues'][0]['tournaments'][0])
```

```python
# https://esports-api.lolesports.com/persisted/gw/getStandings?hl=zh-TW&tournamentId=110423696147088301
latest_teams = []
for tournament in latest_tournaments:
    response = requests.get(f'{host}/persisted/gw/getStandings', params=params | {'tournamentId': tournament['id']}, headers=headers)
    standings = response.json()['data']['standings']
    rankings = standings[0]['stages'][0]['sections'][0]['rankings']
    for ranking in rankings:
        latest_teams.extend(ranking['teams'])
```

```python
  # "https://esports-api.lolesports.com/persisted/gw/getTeams?hl=zh-TW&id=psg-talon" | jq '.data.teams[0].players'
players = []
for team in latest_teams:
    response = requests.get(f'{host}/persisted/gw/getTeams', params=params | {'id': team['slug']}, headers=headers)
    teams = response.json()['data']['teams']
    for _team in teams:
        players.extend(_team['players'] or [])
```

```python
avatars = {}
for player in players:
    name = f'{player["firstName"]}-{player["lastName"]}'.strip().replace(' ', '-')
    avatars[name] = player['image']
    
import os
output_dir = f'../outputs/lolesports'
os.makedirs(output_dir, exist_ok=True)

for name, link in avatars.items():
    response = requests.get(link)
    with open(os.path.join(output_dir, f'{name}.png'), 'wb') as fp:
        fp.write(response.content)
    
```
